```python
"""
EDA on Retail Sales Data - Google Colab ready script

How to use:
1. Open Google Colab -> New Python 3 notebook.
2. Copy the entire contents of this file into the first cell (or upload as a .py and run).
3. Provide your datasets by either mounting your Google Drive, using direct URLs, or dragging files into the Colab file browser.

Notes:
- Replace DATASET_1_PATH and DATASET_2_PATH with your file paths or URLs if you have direct links.
- This script is intentionally defensive: it attempts to detect common column names (e.g., InvoiceDate, CustomerID, Quantity, UnitPrice, Product, etc.).
- If your dataset column names differ, change the mapping section below.

Outputs:
- Summary tables printed to screen
- Plots shown inline
- A CSV with cleaned combined data saved to /content/cleaned_retail_data.csv

"""

# ---------------------- Basic setup ----------------------
# If running in colab, uncomment the mount block below and provide Drive path.

# from google.colab import drive
# drive.mount('/content/drive')

# Option A: If you have direct URLs to CSV, set them here. Otherwise leave as None and use upload.
DATASET_1_PATH = None  # e.g. 'https://example.com/retail_sales1.csv'
DATASET_2_PATH = None  # e.g. 'https://example.com/retail_sales2.csv'

# ---------------------- Install & imports ----------------------
# (colab usually has these; statsmodels might need install)
try:
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt
    import seaborn as sns
    import io
    import os
except Exception as e:
    raise

# Optional: for decomposition
try:
    import statsmodels.api as sm
except Exception:
    # If statsmodels missing, try to install (only in Colab)
    import sys
    !{sys.executable} -m pip install statsmodels --quiet
    import statsmodels.api as sm

sns.set(style="whitegrid")
%matplotlib inline

# ---------------------- Helper functions ----------------------

def try_read(path_or_buffer):
    """Try to read CSV/Excel depending on extension or buffer."""
    if path_or_buffer is None:
        return None
    if isinstance(path_or_buffer, (str,)):
        lower = path_or_buffer.lower()
        if lower.endswith('.csv') or lower.startswith('http'):
            return pd.read_csv(path_or_buffer, low_memory=False)
        if lower.endswith('.xlsx') or lower.endswith('.xls'):
            return pd.read_excel(path_or_buffer)
    # fallback try CSV
    try:
        return pd.read_csv(path_or_buffer, low_memory=False)
    except Exception:
        try:
            return pd.read_excel(path_or_buffer)
        except Exception:
            raise ValueError('Unable to read the provided dataset: %s' % str(path_or_buffer))


def upload_files():
    """If running in Colab and no path provided, this will prompt upload."""
    try:
        from google.colab import files
        uploaded = files.upload()
        # returns first uploaded file path-like buffer for immediate reading
        names = list(uploaded.keys())
        if len(names) == 0:
            return None, None
        # If only one file, return it as dataset1
        if len(names) == 1:
            return names[0], None
        return names[0], names[1]
    except Exception:
        # Not in Colab or files.upload not available
        return None, None

# ---------------------- Data loading ----------------------
print('\nSTEP 1: Load datasets')

df1 = try_read(DATASET_1_PATH) if DATASET_1_PATH else None
df2 = try_read(DATASET_2_PATH) if DATASET_2_PATH else None

if df1 is None and df2 is None:
    print('No dataset paths provided. If you are in Colab, a file picker will open for upload.')
    p1, p2 = upload_files()
    if p1:
        df1 = try_read(p1)
    if p2:
        df2 = try_read(p2)

# If still None, raise
if df1 is None and df2 is None:
    raise RuntimeError('No datasets were provided. Set DATASET_1_PATH/DATASET_2_PATH or upload files.')

# If only one dataset provided, keep df2 as None and proceed
print('Dataset 1 shape:', None if df1 is None else df1.shape)
print('Dataset 2 shape:', None if df2 is None else df2.shape)

# ---------------------- Inspect & unify column names ----------------------
print('\nSTEP 2: Inspect columns and map common names')

def quick_inspect(df, name='Dataset'):
    print(f'\n--- {name} head ---')
    display(df.head())
    print('\nColumns:', list(df.columns)[:50])
    print(df.dtypes)

if df1 is not None:
    quick_inspect(df1, 'Dataset 1')
if df2 is not None:
    quick_inspect(df2, 'Dataset 2')

# Common expected columns and flexible mapping: change these if your names differ
# Typical retail columns: InvoiceNo, InvoiceDate, CustomerID, Country, StockCode/ProductCode, Description/Product, Quantity, UnitPrice, Total/Amount/Sales
col_map = {
    'invoice': ['invoice', 'invoiceno', 'orderid', 'order_id', 'order no', 'billno', 'transaction id'],
    'date': ['invoicedate', 'date', 'orderdate', 'order_date', 'datetime'],
    'customer': ['customerid', 'customer id', 'customer_id', 'cust_id'],
    'product': ['description', 'product', 'productname', 'item', 'stockcode', 'product_code', 'Product Category'],
    'quantity': ['quantity', 'qty', 'units'],
    'price': ['unitprice', 'price', 'unit price', 'amount per item', 'price per unit'],
    'sales': ['total amount', 'amount', 'sales']
}


def map_columns(df):
    new_cols = {}
    lc = {c.lower().strip(): c for c in df.columns}
    for target, candidates in col_map.items():
        for cand in candidates:
            if cand.lower().strip() in lc:
                new_cols[lc[cand.lower().strip()]] = target
                break
    # Rename
    df = df.rename(columns=new_cols)
    return df

if df1 is not None:
    df1 = map_columns(df1)
if df2 is not None:
    df2 = map_columns(df2)

print('\nAfter mapping, Dataset 1 columns (first 30):', list(df1.columns)[:30])
if df2 is not None:
    print('After mapping, Dataset 2 columns (first 30):', list(df2.columns)[:30])

# ---------------------- Basic cleaning ----------------------
print('\nSTEP 3: Basic cleaning')

def basic_clean(df):
    df = df.copy()
    # strip column names
    df.columns = [c.strip() for c in df.columns]
    # parse dates
    if 'date' in df.columns:
        df['date'] = pd.to_datetime(df['date'], errors='coerce')
    # numeric conversions
    if 'quantity' in df.columns:
        df['quantity'] = pd.to_numeric(df['quantity'], errors='coerce')
    if 'price' in df.columns:
        df['price'] = pd.to_numeric(df['price'], errors='coerce')
    # create sales amount if possible, prioritize existing 'sales' column
    if 'sales' not in df.columns and 'quantity' in df.columns and 'price' in df.columns:
         df['sales'] = df['quantity'] * df['price']
    elif 'sales' in df.columns:
         df['sales'] = pd.to_numeric(df['sales'], errors='coerce') # Ensure sales column is numeric

    # remove duplicates
    df = df.drop_duplicates()
    return df

if df1 is not None:
    df1 = basic_clean(df1)
if df2 is not None:
    df2 = basic_clean(df2)

# Combine if both present
if df2 is not None:
    combined = pd.concat([df1, df2], ignore_index=True, sort=False)
else:
    combined = df1.copy()

print('Combined shape:', combined.shape)

# Save intermediate cleaned file
combined.to_csv('/content/cleaned_retail_data_raw.csv', index=False)

# ---------------------- Descriptive statistics ----------------------
print('\nSTEP 4: Descriptive statistics')

print('\nNumeric summary:')
display(combined.describe(include=[np.number]).T)

print('\nCategorical summary (top 10 for each):')
for col in combined.select_dtypes(include=['object', 'category']).columns:
    print('\n--', col, 'top values:')
    print(combined[col].value_counts(dropna=False).head(10))

# Mode example
modes = {}
for c in combined.columns:
    try:
        modes[c] = combined[c].mode().iloc[0]
    except Exception:
        modes[c] = None
print('\nModes for columns (sample):')
for k, v in list(modes.items())[:20]:
    print(k, ':', v)

# ---------------------- Missing data & basic imputation ----------------------
print('\nSTEP 5: Missing data analysis')
missing = combined.isna().mean().sort_values(ascending=False)
print(missing[missing > 0].head(30))

# Strategy: drop rows missing essential fields like date or customer (optional)
essential = []
if 'date' in combined.columns:
    essential.append('date')
if 'sales' in combined.columns:
    essential.append('sales')

print('\nEssential fields detected:', essential)

# ---------------------- Time series analysis ----------------------
print('\nSTEP 6: Time series analysis')
if 'date' in combined.columns and 'sales' in combined.columns: # Check if sales column exists
    ts = combined.dropna(subset=['date']).set_index('date')
    # resample daily and monthly
    daily_sales = ts['sales'].resample('D').sum()
    monthly_sales = ts['sales'].resample('M').sum()

    print('\nPlotting total sales over time (monthly aggregate)')
    plt.figure(figsize=(12,5))
    monthly_sales.plot()
    plt.title('Monthly Sales')
    plt.ylabel('Sales')
    plt.xlabel('Month')
    plt.tight_layout()
    plt.show()

    # seasonal decomposition - requires sufficient length
    try:
        decomposition = sm.tsa.seasonal_decompose(monthly_sales.dropna(), model='additive', period=12)
        fig = decomposition.plot()
        fig.set_size_inches(12,8)
        plt.show()
    except Exception as e:
        print('Could not run seasonal_decompose:', e)

    # Heatmap: month vs year
    df_heat = monthly_sales.reset_index()
    df_heat['year'] = df_heat['date'].dt.year
    df_heat['month'] = df_heat['date'].dt.month
    pivot = df_heat.pivot_table(index='month', columns='year', values='sales', aggfunc='sum')
    plt.figure(figsize=(10,6))
    sns.heatmap(pivot, annot=True, fmt='.0f')
    plt.title('Sales: Month vs Year')
    plt.show()
else:
    print('No date or sales column detected, skipping time series analysis')

# ---------------------- Customer and product analysis ----------------------
print('\nSTEP 7: Customer and product analysis')

# Top products
if 'product' in combined.columns and 'sales' in combined.columns:
    prod = combined.groupby('product').agg({'sales':'sum','quantity':'sum'}).sort_values('sales', ascending=False).reset_index()
    print('\nTop 10 products by sales:')
    display(prod.head(10))
    plt.figure(figsize=(10,6))
    sns.barplot(y='product', x='sales', data=prod.head(10))
    plt.title('Top 10 Products by Sales')
    plt.xlabel('Sales')
    plt.ylabel('Product')
    plt.tight_layout()
    plt.show()
else:
    print('Skipping product analysis: "product" or "sales" column not found.')


# Top customers
if 'customer' in combined.columns and 'sales' in combined.columns:
    cust = combined.groupby('customer').agg({'sales':'sum','invoice':'nunique'}).sort_values('sales', ascending=False).reset_index()
    print('\nTop 10 customers by sales:')
    display(cust.head(10))
    plt.figure(figsize=(10,6))
    sns.barplot(y='customer', x='sales', data=cust.head(10))
    plt.title('Top 10 Customers by Sales')
    plt.xlabel('Sales')
    plt.ylabel('Customer')
    plt.tight_layout()
    plt.show()
else:
    print('Skipping customer analysis: "customer" or "sales" column not found.')

# RFM analysis (simple)
print('\nSTEP 8: Simple RFM analysis (Recency, Frequency, Monetary)')
if 'customer' in combined.columns and 'date' in combined.columns and 'sales' in combined.columns:
    snapshot_date = combined['date'].max() + pd.Timedelta(days=1)
    rfm = combined.groupby('customer').agg({
        'date': lambda x: (snapshot_date - x.max()).days,
        'invoice': 'nunique' if 'invoice' in combined.columns else ('date' if 'date' in combined.columns else 'customer'),
        'sales': 'sum'
    }).reset_index()
    rfm.columns = ['customer','recency','frequency','monetary']
    display(rfm.head())
    # Simple segmentation
    rfm['r_score'] = pd.qcut(rfm['recency'], 4, labels=[4,3,2,1])
    rfm['f_score'] = pd.qcut(rfm['frequency'].rank(method='first'), 4, labels=[1,2,3,4])
    # Use method='first' for monetary score as well
    rfm['m_score'] = pd.qcut(rfm['monetary'].rank(method='first'), 4, labels=[1,2,3,4])
    rfm['RFM_Score'] = rfm['r_score'].astype(str) + rfm['f_score'].astype(str) + rfm['m_score'].astype(str)
    print('\nRFM distribution (top 10 highest monetary):')
    display(rfm.sort_values('monetary', ascending=False).head(10))
else:
    print('Insufficient columns for RFM (need customer, date, sales).')

# ---------------------- Visualization ideas ----------------------
print('\nSTEP 9: Additional visualizations you can run')
print('- Sales by country (if country column exists)')
print('- Sales by hour of day / day of week (if date has time)')
print('- Basket analysis (market basket) to find frequently bought together items')

# ---------------------- Recommendations (automated template) ----------------------
print('\nSTEP 10: Automated recommendations (template)')
recommendations = []
# concentration of sales among top products
if 'product' in combined.columns and 'sales' in combined.columns:
    prod = combined.groupby('product').agg({'sales':'sum','quantity':'sum'}).sort_values('sales', ascending=False).reset_index()
    # Check if prod is empty or sales column has non-numeric values before calculating top_share
    if not prod.empty and pd.api.types.is_numeric_dtype(prod['sales']):
        total_sales = prod['sales'].sum()
        if total_sales > 0:
            top_share = prod.head(10)['sales'].sum() / total_sales
            recommendations.append(f'Top 10 products account for {top_share:.2%} of sales — consider promoting long-tail products or ensuring stock for top sellers.')
        else:
            recommendations.append('Cannot calculate top product sales share as total sales are zero.')
    else:
        recommendations.append('Cannot calculate top product sales share due to missing product or sales data.')
# customer concentration
if 'customer' in combined.columns and 'sales' in combined.columns:
    cust = combined.groupby('customer').agg({'sales':'sum','invoice':'nunique'}).sort_values('sales', ascending=False).reset_index()
    if not cust.empty and pd.api.types.is_numeric_dtype(cust['sales']):
        total_cust_sales = cust['sales'].sum()
        if total_cust_sales > 0:
            top_cust_share = cust.head(10)['sales'].sum() / total_cust_sales
            recommendations.append(f'Top 10 customers account for {top_cust_share:.2%} of sales — consider loyalty or retention strategies.')
        else:
             recommendations.append('Cannot calculate top customer sales share as total sales are zero.')
    else:
        recommendations.append('Cannot calculate top customer sales share due to missing customer or sales data.')
# seasonality
if 'date' in combined.columns and 'sales' in combined.columns:
     # Check if monthly_sales is available before appending recommendation
    try:
        _ = monthly_sales
        recommendations.append('If monthly/seasonal spikes exist, align inventory and promotions to peak months. Use the seasonal decomposition plots to find months with high seasonal components.')
    except NameError:
        pass


print('\nRecommendations:')
for r in recommendations:
    print('- ', r)
if len(recommendations) == 0:
    print('No automated recommendations generated (insufficient data).')

# ---------------------- Save cleaned dataset ----------------------
clean_path = '/content/cleaned_retail_data.csv'
combined.to_csv(clean_path, index=False)
print('\nCleaned combined dataset saved to', clean_path)

print('\nDONE. Modify the column mapping at the top of the script if your dataset uses different names.')
```
